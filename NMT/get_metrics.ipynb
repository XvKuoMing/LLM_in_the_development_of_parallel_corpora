```python
def join_words(word_list):
    sentence = ''
    for word in word_list:
        if word in {'.', '!', ',', '?', '—'}:
            sentence = sentence.rstrip() + word
        else:
            sentence += word + ' '
    return sentence.strip()

ru = [join_words(words) for words in ru]
```

```python
udm = [join_words(words) for words in udm]
ru_pred = [join_words(words) for words in ru_pred]
```
